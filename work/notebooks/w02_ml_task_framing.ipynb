## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*


A fixed rule (like stale × visible) uses hand-picked thresholds that don't adapt to interactions between features. A model can learn combinations — e.g., moderate staleness plus very low CTR — that a human-designed rule would miss, and its Precision@50 can improve as more data comes in, without manually re-tuning thresholds.
